```python
import xlrd
from xlrd import *
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import itertools
%matplotlib inline
```

```python
data=xlrd.open_workbook("C:/Users/user/OneDrive/Work/UB_Master/s_ws_2016_2017/Michael_Hofmann_2017_2nd_lab/data.xlsx")

# List sheet names, and pull a sheet by name
#
sheet_names = data.sheet_names()
print('Sheet Names', sheet_names)
```

```python
table_sheet = data.sheet_by_name(sheet_names[3])
# Or grab the first sheet by index 
    #xl_sheet = xl_workbook.sheet_by_index(0)
# sheet1= excel_sheet.sheet_by_name('parents')

# for i in range(0,table_sheet.nrows):        
#     row= table_sheet.row_slice(i)        
#     NG = row[0].value        
#     NPT = row[1].value        
#     CM = row[2].value
#     IL = row[3].value
# r = sheet.row(0) #returns all the CELLS of row 0,
# c = sheet.col_values(0) #returns all the VALUES of column 0,
NG=table_sheet.row_values(0)
NPT=table_sheet.row_values(1)
CM=table_sheet.row_values(2)
IL=table_sheet.row_values(3)
#now start to whisker box plotting
# ratio = [NG, NPT, CM, IL]
#trying to figure out 1D plot?
name = np.arange(0,len(NG)) # this is the value where you want the data to appear on the y-axis.
#885 29345 3171956 93750 1880 2101630 147905 47157 
#49302 1221203 13120 37162 23207 143816 68326 19201 84618 3592123 3341804 77383 72347 104419
name_p=[[name[0]],[name[1]],name[2:29],name[29:37],name[37],name[38:46],name[46:55],name[55],name[56],name[57:61],name[61],name[62],name[63:66],name[66],name[67:69],name[69:72],name[72],name[73:88],name[88:91],name[91],name[92],name[93:100]]
####################
NG_p=[NG[0],NG[1],NG[2:29],NG[29:37],NG[37],NG[38:46],NG[46:55],NG[55],NG[56],NG[57:61],NG[61],NG[62],NG[63:66],NG[66],NG[67:69],NG[69:72],NG[72],NG[73:88],NG[88:91],NG[91],NG[92],NG[93:100]]
########################
NPT_p=[NPT[0],NPT[1],NPT[2:29],NPT[29:37],NPT[37],NPT[38:46],NPT[46:55],NPT[55],NPT[56],NPT[57:61],NPT[61],NPT[62],NPT[63:66],NPT[66],NPT[67:69],NPT[69:72],NPT[72],NPT[73:88],NPT[88:91],NPT[91],NPT[92],NPT[93:100]]
###########################
CM_p=[CM[0],CM[1],CM[2:29],CM[29:37],CM[37],CM[38:46],CM[46:55],CM[55],CM[56],CM[57:61],CM[61],CM[62],CM[63:66],CM[66],CM[67:69],CM[69:72],CM[72],CM[73:88],CM[88:91],CM[91],CM[92],CM[93:100]]
############################
IL_p=[IL[0],IL[1],IL[2:29],IL[29:37],IL[37],IL[38:46],IL[46:55],IL[55],IL[56],IL[57:61],IL[61],IL[62],IL[63:66],IL[66],IL[67:69],IL[69:72],IL[72],IL[73:88],IL[88:91],IL[91],IL[92],IL[93:100]]

print(len(name_p),len(NG_p),len(NPT_p),len(CM_p),len(IL_p))
```

```python
#categorized by region
alone=plt.figure(figsize=(15,15))
#for colormap
# colors = iter(cm.prism(np.linspace(0, 0.6, len(name_p))))
colors1=itertools.cycle(['#9EC53A','#50BEA3','#38784C','#9B84F0','#C54F71','#783C75','#3399E1','#006078','#052F61','#DE9C3C','#E76F39','#C62324',])
colors2=itertools.cycle(['#9EC53A','#50BEA3','#38784C','#9B84F0','#C54F71','#783C75','#3399E1','#006078','#052F61','#DE9C3C','#E76F39','#C62324',])

x_ax1=[0,0]
x_ax2=[100,100]
y_ng=[0,0.15]
y_npt=[0,0.04]
y_cm=[0,0.08]
y_il=[0.82,0.98]

#NG===================================
labels1=['885','29345','3171956','93750','1880','2101630','147905','47157','49302','1221203','13120','37162','23207','143816','68326','19201','84618','3592123','3341804','77383','72347','104419']
ng_labels=[]
ng=alone.add_subplot(2,2,1)
npt_labels=[]
npt=alone.add_subplot(2,2,2)
cm_labels=[]
cm=alone.add_subplot(2,2,3)
il_labels=[]
il=alone.add_subplot(2,2,4)
plt.subplots_adjust(wspace=0.5, hspace=0.3)
for i in range(0,len(name_p)):
    ng.scatter(name_p[i],NG_p[i],s=30,marker='o',color=next(colors1))
    ng=alone.add_subplot(2,2,1)
    ng_labels.append(labels1[i])


plt.gca().set_prop_cycle(None)
plt.ylim([0,0.15])
plt.xlim([-10,110])
plt.title('NG', fontsize=30)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.legend(ng_labels,bbox_to_anchor=(1.35, 1.02),loc='upper right')
ng.set_axis_bgcolor('#f7f7f7')
ng.plot(x_ax1,y_ng,'k--',x_ax2,y_ng,'k--')
plt.grid()
# #NPT==================================
plt.gca().set_prop_cycle(None)
for i in range(0,len(name_p)):
    npt.scatter(name_p[i],NPT_p[i],s=30,marker='o',color=next(colors2))
    npt=alone.add_subplot(2,2,2)
    npt_labels.append(labels1[i])

plt.ylim([0,0.04])
plt.xlim([-10,110])
plt.legend(npt_labels,bbox_to_anchor=(1.35, 1.02),loc='upper right')
npt.set_axis_bgcolor('#f7f7f7')
plt.title('NPT', fontsize=30)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
npt.plot(x_ax1,y_npt,'k--',x_ax2,y_npt,'k--')
plt.grid()
#CM===================================
for i in range(0,len(name_p)):
    cm.scatter(name_p[i],CM_p[i],s=30,marker='o',color=next(colors1))
    cm=alone.add_subplot(2,2,3)
    cm_labels.append(labels1[i])

plt.ylim([0,0.08])
plt.xlim([-10,110])
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
cm.set_axis_bgcolor('#f7f7f7')
plt.title('CM', fontsize=30)
plt.legend(cm_labels,bbox_to_anchor=(1.35, 1.02),loc='upper right')
cm.plot(x_ax1,y_cm,'k--',x_ax2,y_cm,'k--')
plt.grid()
# #IL===================================
for i in range(0,len(name_p)):
    il.scatter(name_p[i],IL_p[i],s=30,marker='o',color=next(colors2))
    il=alone.add_subplot(2,2,4)
    il_labels.append(labels1[i])
    
plt.ylim([0.82,0.98])
plt.xlim([-10,110])
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
il.set_axis_bgcolor('#f7f7f7')
plt.title('IL', fontsize=30)
plt.legend(il_labels,bbox_to_anchor=(1.35, 1.02),loc='upper right')
il.plot(x_ax1,y_il,'k--',x_ax2,y_il,'k--')
plt.grid()
plt.show()


```
