```python
#categorized by region

new_style = {'grid': True}
sns.set_style("dark")
alone=plt.figure(figsize=(15,15))
#for colormap
# colors = iter(cm.prism(np.linspace(0, 0.6, len(name_p))))
colors1=itertools.cycle(['#9EC53A','#50BEA3','#38784C','#9B84F0','#C54F71','#783C75','#3399E1','#006078','#052F61','#DE9C3C','#E76F39','#C62324',])
colors2=itertools.cycle(['#9EC53A','#50BEA3','#38784C','#9B84F0','#C54F71','#783C75','#3399E1','#006078','#052F61','#DE9C3C','#E76F39','#C62324',])

x_ax1=[0,0]
x_ax2=[100,100]
y_ng=[0,0.15]
y_npt=[0,0.04]
y_cm=[0,0.08]
y_il=[0.82,0.98]

#NG===================================
labels1=['Holocentriformes','Kurtiformes','Gobiiformes','IncertaeSedis','Mugiliformes','Cichliformes','Blenniiformes','Gobiesociformes','Atheriniformes','Cyprinodontiformes','Synbranchiformes','Carangiformes','Anabantiformes','Pleuronectiformes','Syngnathiformes','Callionymiformes','Labriformes','Perciformes','Scorpaeniformes','Acanthuriformes','Lophiiformes','Tetraodontiformes']
ng_labels=[]
ng=alone.add_subplot(2,2,1)
npt_labels=[]
npt=alone.add_subplot(2,2,2)
cm_labels=[]
cm=alone.add_subplot(2,2,3)
il_labels=[]
il=alone.add_subplot(2,2,4)
plt.subplots_adjust(wspace=0.5, hspace=0.3)
for i in range(0,len(name_p)):
    ng.scatter(name_p[i],NG_p[i],s=30,marker='o',color=next(colors1))
    ng=alone.add_subplot(2,2,1)
    ng_labels.append(labels1[i])


plt.gca().set_prop_cycle(None)
plt.ylim([0,0.15])
plt.xlim([-10,110])
plt.title('NG', fontsize=30)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.legend(ng_labels,bbox_to_anchor=(1.35, 1.02),loc='upper right')
ng.plot(x_ax1,y_ng,'k--',x_ax2,y_ng,'k--')
plt.grid()
# #NPT==================================
plt.gca().set_prop_cycle(None)
for i in range(0,len(name_p)):
    npt.scatter(name_p[i],NPT_p[i],s=30,marker='o',color=next(colors2))
    npt=alone.add_subplot(2,2,2)
    npt_labels.append(labels1[i])

plt.ylim([0,0.04])
plt.xlim([-10,110])
plt.legend(npt_labels,bbox_to_anchor=(1.35, 1.02),loc='upper right')
plt.title('NPT', fontsize=30)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
npt.plot(x_ax1,y_npt,'k--',x_ax2,y_npt,'k--')
plt.grid()
#CM===================================
for i in range(0,len(name_p)):
    cm.scatter(name_p[i],CM_p[i],s=30,marker='o',color=next(colors1))
    cm=alone.add_subplot(2,2,3)
    cm_labels.append(labels1[i])

plt.ylim([0,0.08])
plt.xlim([-10,110])
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.title('CM', fontsize=30)
plt.legend(cm_labels,bbox_to_anchor=(1.35, 1.02),loc='upper right')
cm.plot(x_ax1,y_cm,'k--',x_ax2,y_cm,'k--')
plt.grid()
# #IL===================================
for i in range(0,len(name_p)):
    il.scatter(name_p[i],IL_p[i],s=30,marker='o',color=next(colors2))
    il=alone.add_subplot(2,2,4)
    il_labels.append(labels1[i])

plt.ylim([0.82,0.98])
plt.xlim([-10,110])
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.title('IL', fontsize=30)
plt.legend(il_labels,bbox_to_anchor=(1.35, 1.02),loc='upper right')
il.plot(x_ax1,y_il,'k--',x_ax2,y_il,'k--')
plt.grid()
plt.show()


```

```python
#try to use seaborn and pands for strip plot
boxes=plt.figure(figsize=(25,30))
plt.subplots_adjust(wspace=0.5, hspace=0.3)
boxes1=boxes.add_subplot(2,2,1)
boxes2=boxes.add_subplot(2,2,2)
boxes3=boxes.add_subplot(2,2,3)
boxes4=boxes.add_subplot(2,2,4)
dfe = pd.ExcelFile("C:/Users/user/OneDrive/Work/UB_Master/s_ws_2016_2017/Michael_Hofmann_2017_2nd_lab/data.xlsx")
dfs= dfe.parse('snsd')
sns.set_style('darkgrid')
sns.set_context(context='poster',font_scale=1.2)#rc={'font.size':60.0,'axes.titlesize':60.0,'axes.labelsize':40.0})
ngs=sns.stripplot(x='NG',y='Family',data=dfs,size=10,jitter=True, edgecolor="k",linewidth=0.5,ax=boxes1)
npts=sns.stripplot(x='NPT',y='Family',data=dfs,size=10,jitter=True, edgecolor="k",linewidth=0.5,ax=boxes2)
cms=sns.stripplot(x='CM',y='Family',data=dfs,size=10,jitter=True, edgecolor="k",linewidth=0.5,ax=boxes3)
ils=sns.stripplot(x='IL',y='Family',data=dfs,size=10,jitter=True, edgecolor="k",linewidth=0.5,ax=boxes4)
# npts.set(ylabel='',yticks=([]))
ngs.set_xlabel('NG/Hypothalamus',fontsize=40,labelpad=30)
ngs.set_ylabel('Family',fontsize=40,labelpad=30)
npts.set_xlabel('NPT/Hypothalamus',fontsize=40,labelpad=30)
npts.set_ylabel('Family',fontsize=40,labelpad=30)
cms.set_xlabel('CM/Hypothalamus',fontsize=40,labelpad=30)
cms.set_ylabel('Family',fontsize=40,labelpad=30)
ils.set_xlabel('IL/Hypothalamus',fontsize=40,labelpad=30)
ils.set_ylabel('Family',fontsize=40,labelpad=30)
plt.show()
```
